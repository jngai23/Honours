```python
#11/10/24 Unites all model outputs into 2 amalgam datasets, one sorted by feature, other by model type

import pandas as pd
import numpy as np
import re
import os

```

```python
#load in FP results
with open('/Users/james/Documents/Honours/Results/Figure_Development/Fingerprints/modeloutputs/FPNNResults.txt', 'r') as file:
    text = file.read()
pattern = re.compile(r'\bmcc = .+$', re.MULTILINE)
lines = pattern.findall(text)
results = []
for line in lines:
    results.append(float(line[6:]))
    

datasets = []
index = []
directory = '/Users/james/Documents/Honours/Results/Model_Outputs/'
namelist = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if 'csv' in filename:
        pathname = directory + file
        df = pd.read_csv(pathname)
        datasets.append(df)
        namelist.append(filename)
        
index = ['SVM', 'RF', 'KNN', 'NN']
mcclist = []
for data in datasets:
    mcclist.append(data['MCC'].values.tolist())
mcclist.append(results)

datalist = datasets[0]['Dataset'].values
temp = []
for item in datalist:
    temp.append(item[:-13])
datalist = temp

rfdf = pd.DataFrame(mcclist)
rfdf = rfdf.transpose()
rfdf.columns = index
rfdf.index = datalist
rfdf = rfdf.sort_index()
fpdf = rfdf

fpdf.to_csv('/Users/james/Documents/Honours/Results/conglomdata/FPdata.csv')
```

```python
#Dbank

with open('/Users/james/Documents/Honours/Results/Figure_Development/DrugBank Targets/modelout/drugbankNNout.txt', 'r') as file:
    text = file.read()
pattern = re.compile(r'\bmcc = .+$', re.MULTILINE)
lines = pattern.findall(text)
results = []
for line in lines:
    results.append(float(line[6:]))
    
datasets = []
index = []
directory = '/Users/james/Documents/Honours/Results/Figure_Development/DrugBank Targets/modelout/'
namelist = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if 'csv' in filename:
        pathname = directory + file
        df = pd.read_csv(pathname)
        datasets.append(df)
        namelist.append(filename)
        
temp = datasets[0]
datasets[0] = datasets[1]
datasets[1] = temp

temp = namelist[0]
namelist[0] = namelist[1]
namelist[1] = temp

index = ['SVM', 'RF', 'KNN', 'NN']
mcclist = []
for data in datasets:
    mcclist.append(data['MCC'].values.tolist())
mcclist.append(results)

datalist = datasets[0]['Dataset'].values
temp = []
for item in datalist:
    item = re.sub(r'^.*?_', '', item)
    temp.append(item[:-14])
datalist = temp

rfdf = pd.DataFrame(mcclist)
rfdf = rfdf.transpose()
rfdf.columns = index
rfdf.index = datalist
rfdf.loc['Developmental_Toxicity'] = [0, 0, 0, 0]
rfdf.loc['Reproductive_Toxicity'] = [0, 0, 0, 0]
rfdf.loc['Respiratory_Toxicity'] = [0, 0, 0, 0]
rfdf = rfdf.sort_index()
dbdf = rfdf

dbdf.to_csv('/Users/james/Documents/Honours/Results/conglomdata/DrugBankdata.csv')
```
