```python
wholerepdata = [[], [] ,[] ,[]]
loops = 0
for item in dataindex:
    for set in item:
        for val in set:
            wholerepdata[loops].append(val)
    loops += 1
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(wholerepdata[0], wholerepdata[1], wholerepdata[2], wholerepdata[3])
print('ANOVA F-statistic:', f_stat)
print('ANOVA p-value:', p_value)

# Interpret the p-value
if p_value < 0.05:
    print("At least one group mean is significantly different (reject H0).")
else:
    print("No significant difference among group means (fail to reject H0).")
```

```python
import numpy as np
from scipy import stats

# Example data for two independent samples
data1 = np.array(modelvals[2])
data2 = np.array(modelvals[3])

# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(data1, data2)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
```
