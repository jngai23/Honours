```python
result = [s.split(';') for s in mappingdata['Gene Ontology IDs']]
goidlist = []
for gene in result:
    idlist = []
    for string in gene:
        ids = string.replace(" ", "")
        idlist.append(ids)
    goidlist.append(idlist)
```

```python
#Converts the Gene Ontology IDs column into its constituent ids
biglist = []
for string in mappingdata['Gene Ontology IDs']:
    words_list = string.split(';')
    words_list = [word.strip() for word in words_list]
    biglist.append(words_list)
```

```python
#Creates a vocabulary of GOIDs
vocab = []
for idlist in biglist:
    for goid in idlist:
        if goid not in vocab:
            vocab.append(goid)
```

```python
#Creates a 1 hot encoded 2d array of chemblids and goid
featureset = []
for chembl in biglist:
    encodedlist = []
    for word in vocab:
        if word in chembl:
            encodedlist.append(1)
        else:
            encodedlist.append(0)
    
    featureset.append(encodedlist)
```

```python
#Convert array to pandas dataframe with GOids as column names and append original dataframe
columns = []
for word in vocab:
    columns.append(word)
featureframe = pd.DataFrame(featureset, columns=columns)
newframe = pd.concat([mappingdata, featureframe], axis=1)
```
