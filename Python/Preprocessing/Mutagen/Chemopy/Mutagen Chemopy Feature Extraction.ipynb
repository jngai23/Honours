```python
from chemopy import ChemoPy
from rdkit import Chem

molist = []
for smile in smiles:
    molist.append(Chem.MolFromSmiles(smile))
```

```python
molist
```

```python
from chemopy import constitution, topology, connectivity, kappa, basak, estate 
from chemopy import charge, molproperty, moe, geometric, cpsa, morse, rdf

consti_des = []
top_des = []
conec_des = []
kap_des = []
bas_des = []

est_des = []
char_des = []
molprop_des = []
moe_des = []
cpsa_des = []
morse_des = []
rdf_des = []

for mol in molist:
#mol = molist[0]
    con = constitution.Constitution.get_all(mol)
    consti_des.append(con)

    top = topology.Topology.get_all(mol)
    top_des.append(top)

    conec = connectivity.Connectivity.get_all(mol)
    conec_des.append(conec)

    kap = kappa.Kappa.get_all(mol)
    kap_des.append(kap)

    bas = basak.Basak.get_all(mol)
    bas_des.append(bas)

    est = estate.EState.get_all_descriptors(mol)
    est_des.append(est)
    char_des.append(charge.Charge.get_all(mol))
    molprop_des.append(molproperty.MolecularProperties.get_all(mol))
    moe_des.append(moe.MOE.get_all(mol))
```
