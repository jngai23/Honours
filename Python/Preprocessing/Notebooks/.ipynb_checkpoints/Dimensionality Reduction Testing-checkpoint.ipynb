```python
#Autoencoding testing
#using tutorial from https://www.tensorflow.org/tutorials/generative/autoencoder and https://socr.umich.edu/HTML5/ABIDE_Autoencoder/

import tensorflow as tf
import keras
keras.utils.set_random_seed(81)
dataframe
```

```python
## Let's start with latent space with 210 nodes i.e, we are reducing our 3297 feature space to 210 dimensions

num_inputs = 308

## Encoder section
hidden_1 = round(308//1.2)       ## 154 (Number of nodes)
hidden_2 = round(hidden_1 //1.2)  ## 77
hidden_3 = round(hidden_2 //1.2)  ## 33

## Latent Space
hidden_4 = round(hidden_3 //1.2)  ## 16

## Decoder 
hidden_5 = hidden_3      ## 33
hidden_6 = hidden_2      ## 77
hidden_7 = hidden_1      ## 154

num_outputs = num_inputs ## 308

sigmoid = lambda x: 1 / (1 + np.exp(-x))
tanh = lambda x:  (np.exp(x) - np.exp(-x))/ ((np.exp(x) + np.exp(-x)))
def ReLU(x):
    return abs(x) * (x > 0)

x = np.linspace(-10,10,1000)


```

```python
learning_rate = 0.001
activ_fn = tf.nn.relu
tf.compat.v1.disable_eager_execution()
X = tf.compat.v1.placeholder(tf.float32,shape = [None,num_inputs])

np.random.seed(81)
tf.random.set_seed(81)
tf.keras.utils.set_random_seed(81)
initializer = tf.keras.initializers.variance_scaling()

w1 = tf.Variable(initializer([num_inputs,hidden_1]),dtype=tf.float32)
w2 = tf.Variable(initializer([hidden_1,hidden_2]),dtype=tf.float32)
w3 = tf.Variable(initializer([hidden_2,hidden_3]),dtype=tf.float32)
w4 = tf.Variable(initializer([hidden_3,hidden_4]),dtype=tf.float32)
w5 = tf.transpose(w4)
w6 = tf.transpose(w3)
w7 = tf.transpose(w2)
w8 = tf.transpose(w1)

b1 = tf.Variable(tf.zeros(hidden_1))
b2 = tf.Variable(tf.zeros(hidden_2))
b3 = tf.Variable(tf.zeros(hidden_3))
b4 = tf.Variable(tf.zeros(hidden_4))
b5 = b3
b6 = b2
b7 = b1
b8 = tf.Variable(tf.zeros(num_outputs))

```

```python
hid_layer1 = activ_fn(tf.matmul(X,w1)+b1)
hid_layer2 = activ_fn(tf.matmul(hid_layer1,w2)+b2)
hid_layer3 = activ_fn(tf.matmul(hid_layer2,w3)+b3)
hid_layer4 = activ_fn(tf.matmul(hid_layer3,w4)+b4)
hid_layer5 = activ_fn(tf.matmul(hid_layer4,w5)+b5)
hid_layer6 = activ_fn(tf.matmul(hid_layer5,w6)+b6)
hid_layer7 = activ_fn(tf.matmul(hid_layer6,w7)+b7)
output_layer = activ_fn(tf.matmul(hid_layer7,w8)+b8)

loss = tf.reduce_mean(tf.square(output_layer - X))


## Optimizer
optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate)
train = optimizer.minimize(loss)


## Initialzing
init = tf.compat.v1.global_variables_initializer()

saver = tf.compat.v1.train.Saver()

```

```python
num_epochs = 200

batch_size = 32

to_plot_220 = []
from sklearn.utils import shuffle
#X = shuffle(X) 

with tf.compat.v1.Session() as sess:
    sess.run(init)
    for epoch in range(num_epochs):
        for index, offset in enumerate(range(0, dataframe.shape[0], batch_size)):
            xs = dataframe[offset: offset + batch_size]
            #print(offset)
            sess.run(train, feed_dict={X: xs})
        training_loss = loss.eval(feed_dict={X: xs})
        to_plot_220.append(training_loss)
        if epoch % 20 == 0:
          print("Epoch: {} Loss: {}".format(epoch,training_loss))
        
        
    """
    Saving the trained model
    """
    saver.save(sess,'autoencodertest.ckpt')
```

```python
loss = tf.reduce_mean(tf.square(output_layer - X))


l2_regularizer = tf.keras.regularizers.l2(0.005)
weights = tf.compat.v1.trainable_variables()
reg_penalty = tf.reduce_sum([l2_regularizer(w) for w in weights])
reg_loss = loss + reg_penalty

optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate)

train = optimizer.minimize(reg_loss)
```
