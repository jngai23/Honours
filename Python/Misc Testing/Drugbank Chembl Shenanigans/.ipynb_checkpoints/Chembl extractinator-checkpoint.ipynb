```python
#29/8/24 attempts to grab chembl ids from raw toxric datasets
#starting with ames mutagenicity as the canary in the coal mine

import pandas as pd
import numpy as np

#df = pd.read_csv('/Users/james/Documents/Honours/Data/Raws/Mutagenicity_Ames Mutagenicity.csv')
df = pd.read_csv('/Users/james/Downloads/drugbank_all_target_polypeptide_ids.csv/all.csv')
```

```python
idlist = []
for item in df['Drug IDs']:
    temp = item.replace(';', '')
    temp = temp.split()
    idlist.append(temp)
```

```python
uniqlist = []
for item in idlist:
    for drug in item:
        if drug not in uniqlist:
            uniqlist.append(drug)
```

```python
uniqlist
```

```python
import requests
import time
#drug = 'DB07427'
headers = {
    'accept': 'application/json',
    'Content-Type': 'application/json',
}

lnchilist = []
validlist = []

for drug in uniqlist[:25]:
    json_data = {
        'compound': drug,
        'sourceID': 2,
        'type': 'sourceID',
    }

    response = requests.post('https://www.ebi.ac.uk/unichem/api/v1/compounds', headers=headers, json=json_data)
    print(response.status_code)
    while response.status_code != 200:
        print(response.status_code)
        time.sleep(100)
        response = requests.post('https://www.ebi.ac.uk/unichem/api/v1/compounds', headers=headers, json=json_data)
    if len(response.json()['compounds']) != 0:
        lnchilist.append(response.json()['compounds'][0]['standardInchiKey'])
        validlist.append(drug)
    time.sleep(3)
```
