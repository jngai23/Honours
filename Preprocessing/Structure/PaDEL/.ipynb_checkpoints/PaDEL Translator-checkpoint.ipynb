```python
#15/8/24 Given the PaDEL output allocates the SMILES back to their original files

import pandas as pd
import os
import numpy as np

#targetdata = pd.read_csv('/Users/james/Documents/Honours/Data/Unified_SMILES/PaDEL_Prep/PaDEL_Out/padel_unified.csv')
targetdata = pd.read_csv('/Users/james/Documents/Honours/Data/Padel/3D/unified/3dpadelunified.csv')
indexdf = pd.read_csv('/Users/james/Documents/Honours/Data/Unified_SMILES/PaDEL_Prep/padel_index.csv')
index = list(indexdf['0'].values)
```

```python
#Due to how PaDEL reads file inputs, the data is not properly indexed so a reference file is needed
targetdata
```

```python
#First the original SMILES will be reattached to the PaDEL output dataframe in place of its index for clarity
replacements = []
for item in targetdata['Name']:
    item = int(item[8:])
    replacements.append(index[item])
```

```python
#the 'replacements' list also acts as an index for future use
#the 'Name' column is then dropped
targetdata['Name'] = replacements
```

```python
targetdata
```

```python
targetdata = targetdata.fillna(0)
```

```python
datasets = []
index = []
directory = '/Users/james/Documents/Honours/Data/ProcessedRaws/'
namelist = []
smiles = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if 'csv' in filename:
        pathname = directory + file
        df = pd.read_csv(pathname)
        datasets.append(df)
        namelist.append(filename[:-4])
        
framelist = []
for df in datasets:
    smiles = list(df['SMILES'].values)
    temp = pd.DataFrame()
    for smile in smiles:
        loops = 0
        for item in replacements:
            if item == smile:
                temp = pd.concat([temp, targetdata.iloc[loops]], axis = 1)
            loops += 1
    framelist.append(temp)
```

```python
namelist
```

```python
finalist = []
loops = 0
for frame in framelist:
    frame = frame.transpose()
    #frame = frame.loc[:,~frame.columns.duplicated()].copy()
    #frame = frame.loc[~frame.index.duplicated(),:].copy()
    toxvals = (datasets[loops]['Toxicity_Value'].values)
    frame.insert(0, 'Toxicity_Value', toxvals)
    
    finalist.append(frame)
    loops += 1
```

```python
loops = 0
extension = '_PaDEL.csv'
direc = '/Users/james/Documents/Honours/Data/Padel/3D/'
templist = []
for df in finalist:
    newname = namelist[loops] + extension
    path = direc + newname
    
    df.to_csv(path, index = False)
    loops += 1
```
