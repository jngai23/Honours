```python
#First the original SMILES will be reattached to the PaDEL output dataframe in place of its index for clarity
replacements = []
for item in targetdata['Unnamed: 0']:
    replacements.append(index[item])
    
#the 'replacements' list also acts as an index for future use
#the 'Name' column is then dropped
targetdata['Unnamed: 0'] = replacements
```

```python
newdata = targetdata.transpose()
newdata = newdata.iloc[1:]
newdata = newdata.transpose()
```
