```python
#5/8/24 Given a folder extracts Chemopy Features for all datasets inside

from chemopy import constitution, topology, connectivity, kappa, basak, estate 
from chemopy import charge, molproperty, moe, geometric, cpsa, morse, rdf
from chemopy import ChemoPy
from rdkit import Chem
import pandas as pd
import numpy as np
import os

directory = os.fsencode('/Users/james/Documents/Honours/Data/Raws/ProcessedRaws')
dataframes = []
filenames = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)    
    filename = filename
    if 'csv' in filename and 'Respiratory' not in filename:
        length = len(filename)
        datanam = filename[:(length-4)]
        filenames.append(datanam)
        pathname = '/Users/james/Documents/Honours/Data/Raws/ProcessedRaws/' + filename
        newdata = pd.read_csv(pathname)
        dataframes.append(newdata)

loops = 0
path = '/Users/james/Documents/Honours/Data/Chemopy/'
for data in dataframes:
    smiles = data['SMILES']
    toxvals = data['Toxicity_Value']

    featurelist = []
    for mol in smiles:
        mol = Chem.MolFromSmiles(mol)
        features = []
        features.append(constitution.Constitution.get_all(mol))
        features.append(connectivity.Connectivity.get_all(mol))
        features.append(kappa.Kappa.get_all(mol))
        features.append(basak.Basak.get_all(mol))
        features.append(estate.EState.get_all_descriptors(mol))
        features.append(charge.Charge.get_all(mol))
        features.append(molproperty.MolecularProperties.get_all(mol))
        features.append(moe.MOE.get_all(mol))
        featurelist.append(features)

    newframe = []
    for item in featurelist:
        sample = item[0]
        for feature in item[1:]:
            sample = sample | feature
        newframe.append(sample)
    df = pd.DataFrame(toxvals)
    temp = pd.DataFrame(newframe)
    df = pd.concat([df, temp], axis = 1)
    
    extension = '_Chemopy.csv'
    dataname = filenames[loops]
    name = path + dataname + extension
    newframe = pd.DataFrame(newframe)
    newframe.to_csv(name, index=False)
    
    loops += 1
    
    
```
