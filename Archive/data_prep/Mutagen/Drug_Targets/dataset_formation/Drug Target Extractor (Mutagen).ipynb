```python
#15/6/24
#This notebook aims to extract the drug target data taken from the CHEMBL api and convert them to a csv format
#When given a folder of CHEMBLIDs and a list of its corresponding drugs, creates a one hot encoded csv file 
#of each molecule and its respecitve targets

import pandas as pd
import numpy as np
#Data import
dataset = pd.read_csv('/Users/james/Documents/Honours/Data/rawdata/rawmutagendata.csv')
smiles = dataset['SMILES'].tolist()
toxvals = dataset['Toxicity_Value'].tolist()
dflist =  list(zip(smiles, toxvals))
df = pd.DataFrame(dflist, columns = ['SMILES', 'Toxicity_Values'])
```

```python
#Reads in files from a given folder, converting them to a list
from pathlib import Path
path = Path('/Users/james/Documents/Honours/Data/rawdata/mutdtargetdump')

count = 0
CHEMBL_ID_LIST = []
for files in path.iterdir():
    with open(files, 'r') as file:
        chemblids = file.readlines()
    idlist = []
    for ids in chemblids:
        idlist.append(ids.strip())
    
    CHEMBL_ID_LIST.append(idlist)
```

```python
#Creates a list of unique CHEMBLIDs 
sublist = []
for items in CHEMBL_ID_LIST:
    for ids in items:
        sublist.append(ids)

uniqids = set(sublist)
uniqids = list(uniqids)
```

```python
#Creates a 1 hot encoded list of CHEMBLIDs
def onehotencoder(uniqids, inputlist):
    newlist = []
    for ids in uniqids:
        if ids in inputlist:
            newlist.append(1)
        else:
            newlist.append(0)
    return newlist

idlist = []
for items in CHEMBL_ID_LIST:
    idlist.append(onehotencoder(uniqids, items))
    
```

```python
#Converts list into a dataframe and then appends it to existing file
temp = pd.DataFrame(idlist, columns = [uniqids])

targetsdf = pd.concat([df, temp], axis=1)
targetsdf.to_csv('/Users/james/Documents/Honours/Data/Targetdata/mutagen/mutDrugTargets.csv', index=False)
```

```python
#The next few sections help verify that the targets have been properly extracted
newlist = []
for ids in uniqids:
    if ids in CHEMBL_ID_LIST[0]:
        newlist.append(1)
    else:
        newlist.append(0)
```
