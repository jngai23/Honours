```python
#12/6/24 Creation of Morgan Fingerprints for the Endocrine Datasets from Toxric
#Is a complete redux of the file in the other Preprocessing/Endocrine file due to the invalidity of the dataset

import pandas as pd
import numpy as np
import os
#Data import
datasets = []
directory = '/Users/james/Documents/Honours/Data/rawdata/rawendodata/Endodata/processedendodata/'
index = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if 'csv' in filename:
        pathname = directory + file
        df = pd.read_csv(pathname)
        
        smiles = df['SMILES'].tolist()
        toxvals = df['Toxicity_Values'].tolist()
        
        dflist =  list(zip(smiles, toxvals))
        df = pd.DataFrame(dflist, columns = ['SMILES', 'Toxicity_Values'])
        
        datasets.append(df)
        index.append(file)
```

```python
index
```

```python
# Generate MACCS key generation (good for substructures)
from rdkit import Chem
from rdkit.Chem import rdFingerprintGenerator
from rdkit.Chem.Draw import IPythonConsole
from rdkit import DataStructs
import rdkit
from rdkit.Chem import rdMolDescriptors
from rdkit.Chem import MACCSkeys
from rdkit.Avalon import pyAvalonTools
from rdkit.Chem import PandasTools

newframes = []
for df in datasets:
    PandasTools.AddMoleculeColumnToFrame(df,'SMILES', 'Structure')
    Morganfp = []
    for items in df.Structure:
        Morganfp.append(rdMolDescriptors.GetMorganFingerprintAsBitVect(items, radius=2))

    data = []
    loops = 0
    for items in Morganfp:
        items = np.array(items)
        data.append(items)
    Morganframe = pd.DataFrame(data)
    newframe = pd.concat([df, Morganframe], axis=1)

    newframe = newframe.drop('Structure', axis=1)
    newframes.append(newframe)
```

```python
loops = 0
for df in newframes:
    path = '/Users/james/Documents/Honours/Data/structdata/endocrine_redux/morganfp/'
    file = 'morganfp_' + index[loops]
    filename = path + file
    df.to_csv(filename, index=False)
    
    loops = loops + 1
```
