```python
import requests
import re
from io import StringIO
loops = 0
classes = []
GOlist = []
for words in uniprotids:
    loops = loops + 1
    url = "https://www.ebi.ac.uk/QuickGO/services/annotation/downloadSearch?includeFields=goName&withFrom="
    requestURL = (url + words)
    response = requests.get(requestURL, headers={ "Accept" : "text/tsv"})
    if response.status_code == 200:
        temp = StringIO(response.text)
        tempframe = pd.read_csv(temp, delimiter = '\t')

        terms = tempframe['GO TERM']
        uniqterms = list(dict.fromkeys(terms))
        GOlist.append(uniqterms)
    else:
        print(response.status_code)
        print(loops)
```
